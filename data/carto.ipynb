```python
def carto_points (df):
    return gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df["longitud"], df["latitud"]))
```

```python
x = carto_points (equip)
```

```python
def mapa (df):
    for i, rows in df.iterrows():
        if rows['sub_category'] == "schools":
            x = Layer(df, basic_style(size=15, opacity=10, color="#008080"),legends=default_legend('schools'))
        if rows['sub_category'] == "centro_civico":
            y = Layer(df, basic_style(size=15, color="#b4c8a8"), legends=default_legend("Startups"))
        elif rows['sub_category'] == "hospitals":
            z = Layer(df, basic_style(size=15, color="#70a494"), legends=default_legend("Starbucks"))

    return Map([x, y, z], layer_selector=True)
```
