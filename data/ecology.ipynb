```python
# get the mean 
yyyy['total'] = yyyy[['marzo_2023', 'feb_2023', "ene_2023", "mar_2022", "abr_2022", "jun_2022", "jul_2022", "ago_2022", "sep_2022", "oct_2022", "nov_2022", "dic_2022"]].mean(axis=1)
```

```python
yyyy["contaminant"] = yyyy["contaminant"].apply(lambda x: x.strip())
```

```python
air = yyyy.drop(yyyy.loc[yyyy["contaminant"]=="Biomassa Black Carbon"].index)
air = air.drop(air.loc[air["contaminant"]=="Black Carbon"].index)
air = air.drop(air.loc[air["contaminant"]=="Flow 1 (Mesura interna equip)"].index)
air = air.drop(air.loc[air["contaminant"]=="Flow 2 (Mesura interna equip)"].index)
air = air.drop(air.loc[air["contaminant"]=="Flow C (Mesura interna equip)"].index)
```

```python
air = air[["district", "total"]].groupby(by="district").mean()
air.reset_index()
# pd area districtis
dist = pd.read_csv("D:\ironhack\proyectos\Barcelona_UrbanAnalysis\csv\\area_districts.csv")
dist.columns = dist.columns.str.lower()
total_air = dist.merge(air, on="district", how="left")
total_air["total_km"] =  total_air["total"].astype(float) / total_air["area_(km²)"].astype(float)
```

### capacidad acustica

###### De 55 a 75 dB

###### Es un nivel de ruido considerable. Por ejemplo, un aspirador genera 65 dB. Una calle con mucho tráfico alcanza los 75 dB. El despertador o la televisión a un volumen elevado, pueden llegar a los 75 dB, igual que una lavadora, el teléfono móvil o una batidora.

###### De 75 dB a 100 dB

###### Es un nivel alto de ruido, la sensación es molesta. Por ejemplo, en un atasco hay 90 dB de ruido. También el que suele emitir una sirena de policía.
